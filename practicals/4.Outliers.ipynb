### What is the Morley Data Set?

The experiment compared the speed of light in perpendicular directions in an attempt to detect the relative motion of matter through the stationary luminiferous aether ("aether wind"). The result was negative, in that Michelson and Morley found no significant difference between the speed of light in the direction of movement through the presumed aether, and the speed at right angles. This result is generally considered to be the first strong evidence against the then-prevalent aether theory, as well as initiating a line of research that eventually led to special relativity, which rules out a stationary aether.[1]

## Exercise 2

Create box plots for all of the numerical variables in Fisher's Iris Data Set.

Firstly I downloaded the dataset from this link and I added it into the data folder in the repository - https://gist.github.com/curran/a08a1080b88344b0c8a7#file-iris-csv

```python
df = pd.read_csv(r'C:\Users\Kenne\OneDrive\Desktop\Programming\funddata-assessment\data\iris.csv')
## imports the excel sheet and then prints out full sheet.

df.columns = ['sepal_length', 'sepal_width' , 'petal_length', 'petal_width', 'species']

## sets the names of the columns 

print(df.to_string()) 
```

```python
df.info()

## This gives us general info on the dataset.

## The sepal_length, sepal_length, petal_length and petal_width are all "float64" whilset species is an object
```

```python
plt.figure(figsize = (10, 7))
plt.title("Box Plot of Iris Data Set and the 4 variables")
df.boxplot()
```

This gives us general info on the dataset visually using a graph.

Petal_length is the widest margin with the majority of the dataset between between 1.8 - 5.1 cm and the petal width being the smalllest, being between 2.9 and 3.1.

I decided to look at each flowers individually and to compare their sepal lengths and widths and also their petal length and widths.

```python
## Based idea for this on https://www.r-bloggers.com/2015/08/use-box-plots-to-assess-the-distribution-and-to-identify-the-outliers-in-your-dataset/

sns.set(style="whitegrid")

title="Distributions of Sepal Length between the Flowers"

sns.boxplot(x="species", y="sepal_length", data=df)

# increasing font size
plt.title(title, fontsize=26)
# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")

title="Distributions of Sepal Width between the Flowers"

sns.boxplot(x="species", y="sepal_width", data=df)

# increasing font size
plt.title(title, fontsize=26)
# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")

title="Distributions of Petal Width between the Flowers"

sns.boxplot(x="species", y="petal_width", data=df)

# increasing font size
plt.title(title, fontsize=26)
# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")

title="Distributions of Petal Length between the Flowers"

sns.boxplot(x="species", y="petal_length", data=df)

# increasing font size
plt.title(title, fontsize=26)
# Show the plot
plt.show()
```

### Observations from the box plots

* The petal length boxplot shows us that Setosa is the smallest between 1.4 cm and 1.6 cm. Whilst Virginica is the biggest between 5.1 cm and 5.9 cm.

* One key thing that I spotted in general from the box plots is that the petal width and petal length is the smallest with setosa, but the sepal width is the largest in comparison to the other flowers.

* This boxplot gives us an excellent visual of the three species and their sepal length of the 3 species. It shows us that virginica has the biggest sepal length between 6.4 and 6.9 whereas, with setosa as the smallest between 4.8 and 5.2 of sepal length.

* The petal length in general is the smallest value for the 3 species, with setosa being much smaller than the other three.
